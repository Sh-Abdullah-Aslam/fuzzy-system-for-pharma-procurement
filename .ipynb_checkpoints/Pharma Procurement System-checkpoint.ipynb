```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline
```

# Input

## universe of discourse for input variables

```python
x_stock = x_budget = x_demand = np.arange(1, 11)
#x_budget = np.arange(0, 1.1, 0.1)
```

## membership functions

### Stock

```python
low_stock = fuzz.trimf(x_stock, [0, 1, 3])
medium_stock = fuzz.trimf(x_stock, [2, 5, 8])
high_stock = fuzz.trimf(x_stock, [7, 10, 10])
```

### Budget

```python
low_budget = fuzz.sigmf(x_stock, 2, -4)
medium_budget = fuzz.gaussmf(x_budget, 5, 1)
high_budget = fuzz.sigmf(x_stock, 8, 4)
```

### Demand

```python
low_demand = fuzz.trimf(x_demand, [0, 1, 3])
medium_demand = fuzz.trimf(x_demand, [2, 5, 8])
high_demand = fuzz.trimf(x_demand, [7, 10, 10])
```

## Creating figures to show membership functions

```python
_, (stock_axis, budget_axis, demand_axis) = plt.subplots(nrows=3, figsize=(8, 9))

# adding title to axes
stock_axis.set_title('Stock')
budget_axis.set_title('Budget')
demand_axis.set_title('Demand')

# adding membership functions to their respective axes

 ## stock axis
stock_axis.plot(x_stock, low_stock, 'g', label='Low')
stock_axis.plot(x_stock, medium_stock, 'r', label='Medium')
stock_axis.plot(x_stock, high_stock, 'b', label='High')

 ## budget axis
budget_axis.plot(x_budget, low_budget, 'g', label='Low')
budget_axis.plot(x_budget, medium_budget, 'r', label='Medium')
budget_axis.plot(x_budget, high_budget, 'b', label='High')

 ## demand axis
demand_axis.plot(x_demand, low_demand, 'g', label='Low')
demand_axis.plot(x_demand, medium_demand, 'r', label='Medium')
demand_axis.plot(x_demand, high_demand, 'b', label='High')

# adding legends to axes
stock_axis.legend()
budget_axis.legend()
demand_axis.legend()

plt.tight_layout()
```

# Output

## universe of discourse for output

```python
x_order = np.arange(1, 11)
```

## membership function

```python
order_none = fuzz.trapmf(x_order, [0, 0, 10, 10])
order_some = fuzz.trimf(x_stock, [1, 3, 5])
order_half = fuzz.trimf(x_stock, [4, 5, 6])
order_more_than_half = fuzz.trimf(x_stock, [5, 7, 9])
order_all = fuzz.trimf(x_order, [0, 0.1, 0.2])
```

```python
_, (order_axis) = plt.subplots(nrows=1, figsize=(8, 3))

order_axis.plot(x_order, order_none, 'r', label='None')
order_axis.plot(x_order, order_some, 'b', label='Some')
order_axis.plot(x_order, order_half, 'g', label='Half')
order_axis.plot(x_order, order_more_than_half, 'y', label='> half')
order_axis.plot(x_order, order_all, color='pink', label='All')
```

# User inputs for inferencing

```python
stock_input = 6.5
budget_input = 6.5
demand_input = 6.5

stock_level_low = fuzz.interp_membership(x_stock, low_stock, stock_input)
stock_level_medium = fuzz.interp_membership(x_stock, medium_stock, stock_input)
stock_level_high = fuzz.interp_membership(x_stock, high_stock, stock_input)

budget_level_low = fuzz.interp_membership(x_budget, low_budget, budget_input)
budget_level_medium = fuzz.interp_membership(x_budget, medium_budget, budget_input)
budget_level_high = fuzz.interp_membership(x_budget, high_budget, budget_input)

demand_level_low = fuzz.interp_membership(x_demand, low_demand, demand_input)
demand_level_medium = fuzz.interp_membership(x_demand, medium_demand, demand_input)
demand_level_high = fuzz.interp_membership(x_demand, high_demand, demand_input)
```

# Rule base

```python
rule_1 = np.fmin(stock_level_low, demand_level_low)
rule_1 = np.fmin(rule_1, budget_level_low)
rule_2 = np.fmin(stock_level_low, demand_level_low)
rule_2 = np.fmin(rule_2, budget_level_medium)
rule_3 = np.fmin(stock_level_low, demand_level_low)
rule_3 = np.fmin(rule_3, budget_level_high)

rule_4 = np.fmin(stock_level_low, demand_level_medium)
rule_4 = np.fmin(rule_4, budget_level_low)
rule_5 = np.fmin(stock_level_low, demand_level_medium)
rule_5 = np.fmin(rule_5, budget_level_medium)
rule_6 = np.fmin(stock_level_low, demand_level_medium)
rule_6 = np.fmin(rule_6, budget_level_high)

rule_7 = np.fmin(stock_level_low, demand_level_high)
rule_7 = np.fmin(rule_7, budget_level_low)
rule_8 = np.fmin(stock_level_low, demand_level_high)
rule_8 = np.fmin(rule_8, budget_level_medium)
rule_9 = np.fmin(stock_level_low, demand_level_high)
rule_9 = np.fmin(rule_9, budget_level_high)

rule_10 = np.fmin(stock_level_medium, demand_level_low)
rule_10 = np.fmin(rule_10, budget_level_low)
rule_11 = np.fmin(stock_level_medium, demand_level_low)
rule_11 = np.fmin(rule_11, budget_level_medium)
rule_12 = np.fmin(stock_level_medium, demand_level_low)
rule_12 = np.fmin(rule_12, budget_level_high)

rule_13 = np.fmin(stock_level_medium, demand_level_medium)
rule_13 = np.fmin(rule_13, budget_level_low)
rule_14 = np.fmin(stock_level_medium, demand_level_medium)
rule_14 = np.fmin(rule_14, budget_level_medium)
rule_15 = np.fmin(stock_level_medium, demand_level_medium)
rule_15 = np.fmin(rule_15, budget_level_high)

rule_16 = np.fmin(stock_level_medium, demand_level_high)
rule_16 = np.fmin(rule_16, budget_level_low)
rule_17 = np.fmin(stock_level_medium, demand_level_high)
rule_17 = np.fmin(rule_17, budget_level_medium)
rule_18 = np.fmin(stock_level_medium, demand_level_high)
rule_18 = np.fmin(rule_18, budget_level_high)

rule_19 = stock_level_high

# none as output
activation_none = np.fmin(rule_1, rule_10)
activation_none = np.fmin(activation_none, rule_11)
activation_none = np.fmin(activation_none, rule_13)
activation_none = np.fmin(activation_none, rule_19)

# some as output
activation_some = np.fmin(rule_2, rule_4)
activation_some = np.fmin(activation_some, rule_7)
activation_some = np.fmin(activation_some, rule_12)
activation_some = np.fmin(activation_some, rule_14)
activation_some = np.fmin(activation_some, rule_16)
activation_some = np.fmin(activation_some, order_half)

# half as output
activation_half = np.fmin(rule_3, rule_5)
activation_half = np.fmin(activation_half, rule_6)
activation_half = np.fmin(activation_half, rule_8)
activation_half = np.fmin(activation_half, order_half)

# more_than_half as output
activation_more_than_half = np.fmin(rule_6, rule_16)
activation_more_than_half = np.fmin(activation_more_than_half, rule_18)
activation_more_than_half = np.fmin(activation_more_than_half, order_more_than_half)

# all as output
activation_all = np.fmin(rule_9, rule_18)
activation_all = np.fmin(activation_all, order_all)
```

```python
order0 = np.zeros_like(x_order)

fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.fill_between(x_order, order0, activation_none, facecolor='b', alpha=0.7)
ax0.plot(x_order, order_none, 'b', linewidth=0.5, linestyle='--', )

ax0.fill_between(x_order, order0, activation_some, facecolor='g', alpha=0.7)
ax0.plot(x_order, order_some, 'g', linewidth=0.5, linestyle='--')

ax0.fill_between(x_order, order0, activation_half, facecolor='r', alpha=0.7)
ax0.plot(x_order, order_half, 'r', linewidth=0.5, linestyle='--')

ax0.fill_between(x_order, order0, activation_more_than_half, facecolor='r', alpha=0.7)
ax0.plot(x_order, order_more_than_half, 'o', linewidth=0.5, linestyle='--')

ax0.fill_between(x_order, order0, activation_all, facecolor='r', alpha=0.7)
ax0.plot(x_order, order_all, 'y', linewidth=0.5, linestyle='--')

ax0.set_title('Output membership activity')


plt.tight_layout()
```
